Имена столбцов - их интерпретация:
-------------------------------------------------------
- id 		    - Уникальный идентификатор для каждого проданного дома

- yr_built	    - Год постройки дома
- yr_renovated  - Год последнего ремонта дома
- zipcode	    - Почтовый индекс местонахождения дома
- lat		    - Lattitudes (Широта)
- long		    - Longitudes (Долгота)
- sqft_living15 - Жилая площадь домов ближайших 15 соседей (в кв.футах)
- sqft_lot15	- Общая площадь ближайших 15 соседей (в кв.футах)

### 1.1 Создать новый признак price_per_sq_lot, который будет содержать среднюю стоимость за один кв. фут общей площади

```python
# Создадим признак
df['price_per_sq_lot'] = 0
df.iloc[:5, [0, 1, 2, -3, -2, -1]]
```

```python
# Внесем в созданный столбец стоимость кв.фута площади
df['price_per_sq_lot'] = df.apply(lambda rec: (rec.price / rec.sqft_lot if rec.price else 0), axis=1)
df[['id', 'price', 'sqft_lot', 'price_per_sq_lot']].head(10)
```

### 1.2 Создать новый признак "delta_renovated", который будет содержать разницу в годах между годом реновацией дома и годом постройки дома. Если реновации дома не было, то в новом признаке поставьте 0

```python
df['delta_renovated'] = df.apply(lambda rec: (rec.yr_renovated - rec.yr_built if rec.yr_renovated else 0), axis=1)
df[['id', 'yr_renovated', 'yr_built', 'delta_renovated']].head()
```

### 1.3 Создайте признаки года продажи, месяца продажи

```python
df['year_sale'] = 0
df['month_sale'] = 0
df[['id', 'year_sale', 'month_sale']].head()
```

## Задача 3
Составьте несколько сводных таблиц

### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален. Отсортируйте от меньшей стоимости к большей

```python
df_bedrooms = df.groupby('bedrooms'). \
                 agg({'price': ['mean']}).apply(np.floor). \
                 sort_values(by=('price','mean'))
df_bedrooms
```

### 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df_condition = df.groupby('condition'). \
                  agg({'price': ['mean', 'max', 'min']}).apply(np.floor)
df_condition
```

### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
# Вариант-1 с использованием агрегирования
df_view = df.groupby(['waterfront', 'view']).size()
df_view
```

```python
# Вариант-2 с использованием pivot - таблиц 
df.pivot_table(index='view',
                  columns='waterfront', 
                  values='id',
                  aggfunc='count',
                  margins=True,
                  margins_name='Total')
```

### 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
df_floors = df.pivot_table(index='bedrooms',
                           columns='floors', 
                           values='id',
                           aggfunc='count')
print(f'Максимальное число домов по этажности:\n{df_floors.max()}')
df_floors
```

### 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
df_med_price = df.pivot_table(index='condition',
                           columns='grade', 
                           values='price',
                           aggfunc='median')
df_med_price
```
